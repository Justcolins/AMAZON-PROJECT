```python
# Import the required libraries
import sqlite3   
import pandas as pd   
import numpy as np   
import seaborn as sns   
import matplotlib.pyplot as plt   
%matplotlib inline
import zipfile
import os
```

```python
from zipfile import ZipFile
with ZipFile('im.db.zip', 'r') as f:
    f.extractall()
```

```python
# Establish a connection to an SQLite database named 'im.db'.

conn = sqlite3.connect('im.db')
curr = conn.cursor()
```

```python
# Execute an SQL query to retrieve the names of all tables 
curr.execute("""
SELECT name
FROM sqlite_master
WHERE type = 'table';
""")
# Fetch all the results from the executed query.
table_names = curr.fetchall()
table_names
```

```python
#Analyse the columns in movie_basics
pd.read_sql("""
SELECT *
FROM movie_basics

""", conn)
```

```python
#Define SQL Query
q ="""
SELECT genres, AVG(averagerating) AS Genre_avg_rating, SUM(numvotes) AS Num_of_votes, COUNT(*) AS num_entries
FROM movie_basics 
JOIN movie_ratings 
    USING (movie_id)
WHERE start_year < 2023 and genres IS NOT NULL
GROUP BY genres 
HAVING num_entries >= 400
ORDER BY Genre_avg_rating DESC

;

"""
# Execute the SQL query and read the results into a DataFrame using pd.read_sql()
genre_rating_df = pd.read_sql(q, conn)
genre_rating_df
```

```python
popular_rating_df = genre_rating_df.sort_values(['Num_of_votes'],ascending = False)
popular_rating_df.head(5)
```

```python
# Create a Matplotlib figure with a specific size
plt.figure(figsize=(10, 6))

# Plotting the bar chart using Matplotlib
ax = plt.barh(popular_rating_df['genres'], popular_rating_df['Num_of_votes'], color='#1f78b4')

# Set axis labels and title
plt.xlabel('Number of Votes in Millions')
plt.ylabel('Genres')
plt.title('Popular Genre by Number of Votes')

# Show the plot
plt.show()
```

```python
# Create a Matplotlib figure with a specific size
plt.figure(figsize=(10, 6))

# Plotting the bar chart using Matplotlib
ax = plt.barh(popular_rating_df['genres'], popular_rating_df['Genre_avg_rating'], color='#1f78b4')

# Set axis labels and title
plt.xlabel('Number of Votes in Millions')
plt.ylabel('Genres')
plt.title('Popular Genre by average rating')

# Show the plot
plt.show()
```

# BOX OFFICE MOJO

```python
# Read the CSV file 'bom.movie_gross.csv' into a DataFrame and store it as 'box_office'
box_office = pd.read_csv('bom.movie_gross.csv')

# Display the first 10 rows of the 'box_office' DataFrame
box_office[:10]
```

```python
# Display information about the 'box_office' DataFrame 
box_office.info()
```

```python
# Check for missing values in the 'box_office' DataFrame 
box_office.isnull().any()
```

```python
#Drop missing values in the columns
box_office = box_office.dropna()
box_office.info()
```

```python
# Replace commas in the 'foreign_gross' column with an empty string,
# convert the result to a float, and assign it back to the 'foreign_gross' column
box_office['foreign_gross'] = box_office['foreign_gross'].str.replace(',', '').astype(float)

# Display information about the 'box_office' 
box_office.info()

```

```python
#Create a new column called international gross
box_office['international_gross'] = box_office['foreign_gross'] + box_office['domestic_gross']
box_office.head()
```

```python
#Find the mean of the domestic gross
domestic_mean = box_office['domestic_gross'].mean()
domestic_mean
```

```python
#Find the mean of the foreign market
foreign_mean = box_office ['foreign_gross'].mean()
foreign_mean
```

```python
#Find the mean of the international gross
international_mean = box_office['international_gross'].mean()
international_mean
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
market = ["Domestic", "Foreign", "International"]
mean_gross = [47019840.20179372,75790384.84130543,122810225.04309915]
# Plot vertical bars of fixed width by passing x and height values to .bar() function 
ax.bar(market, mean_gross, edgecolor= 'black', color = '#1f78b4')

# Give a title to the bar graph and label the axes
ax.set_title("AVERAGE OF DOMESTIC AND INTERNATIONAL GROSS")
ax.set_xlabel ('MARKETS')
ax.set_ylabel ('GROSS PROFIT IN DOLLARS')
```

# DATA VISUALIZATION FOR STUDIOS GROSS REVENUE

```python
# Calculate the upper quantile (85th quantile) of the 'domestic_gross' column in the 'box_office' DataFrame
upper_quantile = box_office['domestic_gross'].quantile(0.85)

# Select films with 'domestic_gross' above the calculated upper quantile
domestic_films = box_office.loc[box_office['domestic_gross'] > upper_quantile]
```

```python
# Calculate the upper quantile (85th quantile) of the 'international_gross' column in the 'box_office' DataFrame
upper_quantiles = box_office['international_gross'].quantile(0.85)

# Select films with 'international_gross' above the calculated upper quantile
international_films = box_office.loc[box_office['domestic_gross'] > upper_quantiles]
```

```python
# Set Seaborn style 
sns.set_style('whitegrid')

# Create a figure with a specific size
plt.figure(figsize=(10, 6))

# Plotting the bar chart
labels = domestic_films.groupby("studio")['domestic_gross'].agg('sum').sort_values().reset_index()
ax = sns.barplot(x='studio', y='domestic_gross', data= labels, color="#1f78b4")

# Set axis labels and title
ax.set_xlabel('Studio')
ax.set_ylabel('Domestic Gross')
ax.set_title('Domestic Gross by Studio')

# Rotate x-axis labels for better visibility
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Show the plot
plt.show()
```

```python
# Set Seaborn style 
sns.set_style('whitegrid')

# Create a figure with a specific size
plt.figure(figsize=(10, 6))

# Plotting the horizontal bar chart
labels = domestic_films.groupby("studio")['international_gross'].agg('sum').sort_values().reset_index()
ax = sns.barplot(x='international_gross', y='studio', data= labels, color="#1f78b4", orient='h')

# Set axis labels and title
ax.set_xlabel('International Gross')
ax.set_ylabel('Studio')
ax.set_title('International Gross by Studio (Horizontal Bar Plot)')

# Show the plot
plt.show()
```
